```python
import pandas as pd
```

1. Найти топ-3 дорогих заказа

```python
items = pd.read_csv('olist_order_items_dataset.csv')
items.head()
```

```python
items['total_cost'] = items['price'] + items['freight_value']
items['total_cost'].head()
```

```python
items.sort_values(by='total_cost', ascending=False).head(3)
```

```python
top_3_expensive = items.sort_values(by='total_cost', ascending=False).head(3)
top_3_expensive.head()
```

Посчитать количество заказов по статусу

```python
orders = pd.read_csv('olist_orders_dataset.csv')
#1. Самый частый способ быстро глянуть статистику.
orders['order_status'].value_counts()

#2. Официальный путь (Groupby):
#Этот способ полезен, если тебе нужно посчитать 
#сразу несколько вещей (и среднее, и сумму, и количество).
orders.groupby('order_status').size().sort_values(ascending = False)
```
